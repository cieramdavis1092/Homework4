```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Define the range for k (from 1 to 30)
k_range = range(1, 31)
mean_scores = []

# Loop over k values for hyperparameter tuning
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    # Perform 10-fold cross-validation and calculate mean score
    scores = cross_val_score(knn, X, y, cv=10)
    mean_scores.append(np.mean(scores))

# Plot the mean accuracy for each k
plt.plot(k_range, mean_scores, marker='o')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Mean Accuracy')
plt.title('KNN Hyperparameter Tuning (k=1 to k=30) with 10-Fold Cross-Validation')
plt.grid(True)
plt.show()

# Output the optimal k and its mean accuracy score
optimal_k = k_range[np.argmax(mean_scores)]
optimal_score = np.max(mean_scores)

print(f"Optimal k: {optimal_k}")
print(f"Mean Accuracy for optimal k: {optimal_score:.4f}")
```
