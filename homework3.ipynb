```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Define the range for k (from 1 to 31)
k_range = range(1, 32)
mean_scores = []

# Loop over k values for hyperparameter tuning
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    # Perform 10-fold cross-validation and calculate mean score
    scores = cross_val_score(knn, X, y, cv=10)
    mean_scores.append(np.mean(scores))

# Plot the 'Value of K for KNN' vs. 'Cross-Validated Accuracy'
plt.plot(k_range, mean_scores, marker='o', color='b', linestyle='-', linewidth=2)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Value of K for KNN vs Cross-Validated Accuracy')
plt.grid(True)
plt.show()

# Output the optimal k and its mean accuracy score
optimal_k = k_range[np.argmax(mean_scores)]
optimal_score = np.max(mean_scores)

print(f"Optimal k: {optimal_k}")
print(f"Mean Accuracy for optimal k: {optimal_score:.4f}")
```
